```
tweet = df['Tweet'] #ambil kolom text
# Menghapus link
tweetnolink = []
for line in tweet:
    result = re.sub(r"http\S+","",line)
    tweetnolink.append(result)
# Menghapus Simbol Retweet
tweetnort = []
for line in tweetnolink:
    result = re.sub(r"RT","",line)
    tweetnort.append(result)
# Menghapus Username
tweetnousername = []
for line in tweetnort:
    result = re.sub(r"@\S+","",line)
    tweetnousername.append(result)
# Case Folding
tweet_lower = []
for line in tweetnousername:
    result = line.lower()
    tweet_lower.append(result)
# Menghapus tagger
tweetnotagger = []
for line in tweet_lower:
  result = re.sub(r'#([^\s]+)',r'\1',line)
  tweetnotagger.append(result)
# Menghapus angka dan angka yang ada dalam string
tweetnonumber = []
for line in tweetnotagger:
  result = re.sub(r'\w*\d\w*','',line).strip()
  tweetnonumber.append(result)
# Menghapus tanda baca
tweetnopunc = []
for line in tweetnonumber:
  result = line.translate(str.maketrans("","",string.punctuation))
  tweetnopunc.append(result)
#Remove additional white spaces
tweetnospace = []
for line in tweetnopunc:
    result = re.sub('[\s]+', ' ', line)
    tweetnospace.append(result)
#Replace #word with word
tweetnodouble = []
for line in tweetnospace:
    result = re.sub(r'#([^\s]+)', r'\1', line)
    tweetnodouble.append(result)
```

```
from __future__ import division
from string import punctuation
positive_word = []
negative_word = []
positive_counts = []
negative_counts = []
clean_tweets = []
header = [("tweets","positif","presentase_pos","presentase_neg")]

tweets = open("tweet.csv").read()
tweet_list = tweets.split('\n')

pos_sent = open("positif.txt").read()
positive_words = pos_sent.split('\n')

neg_sent = open("negatif.txt").read()
negative_words = neg_sent.split('\n')

for tweet in tweet_list:
  text_positif=[]
  text_negatif=[]
  positive_counter = 0
  negative_counter = 0
  tweet_procces_2 = tweet.replace('!','').replace('.', '')
  tweet_processed = tweet_procces_2.lower()

  for p in list (punctuation):
    tweet_processed = tweet_processed.replace(p, '')

  clean_tweets.append(tweet_processed)

  words = tweet_processed.split(' ')
  word_count = len(words)
  for word in words:
    if word in positive_words:
      text_positif.append(word)
      positive_counter = positive_counter+1
    elif word in negative_words:
      text_negatif.append(word)
      negative_counter = negative_counter+1

  positive_counts.append((positive_counter / word_count)*100)
  negative_counts.append((negative_counter / word_count)*100)
  positive_word.append(text_positif)
  negative_word.append(text_negatif)
```

```
df = pd.DataFrame(zip(clean_tweets, positive_word, negative_word, positive_counts, negative_counts))

df.to_csv (r'labelll.csv', index = False)
```
